```python
import requests
import json

webhook_url = "YOUR WEBHOOK URL"

df = spark.sql("""
SELECT
COUNT(*) as total,
SUM(CASE WHEN sensor_status='CRITICAL' THEN 1 ELSE 0 END) as critical_count
FROM iot_sensor_anomoly_detection.gold.fact_device_monitoring
""")

data = df.collect()[0]

total = data["total"]
critical = data["critical_count"]

rate = critical / total if total > 0 else 0

message = {
"text": f"""
🚨 IoT Sensor Alert

Total Records: {total}
Critical Sensors: {critical}
Anomaly Rate: {round(rate*100,2)}%
"""
}

requests.post(webhook_url, data=json.dumps(message))

print("Slack message sent")
```
